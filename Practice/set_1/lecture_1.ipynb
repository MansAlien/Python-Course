```python
# Variables
num_1 = 5
num_2 = 10 
```

```python
# Addition (Integer/Float)
print(num_1+num_2)
```

```python
# Multiplication
print(num_1*num_2)
```

```python
# Addition (String)
print("num_1"+"num_2")
```

```python
# Addition (String)
num_1 = str(num_1)
num_2 = str(num_2)
print(num_1+num_2)
```

```python
name = input('What is your name?')
```

```python
# Addition strings with ","
print(f"my name is", name)
```

```python
# Addition strings with "+"
print(f"my name is " + name)
```
